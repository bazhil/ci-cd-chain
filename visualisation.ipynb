```python
# Визуализация данных
from matplotlib import pyplot as plt

# простой график роста ВВП
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gpg = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# создаю линейную диаграмму: годы по оси Х, ВВП по оси У
plt.plot(years, gpg, color='green', marker='o', linestyle='solid')
# добавляю название диаграммы
plt.title('Номинальный ВВП')
# Добавляю подпись к оси Х
plt.ylabel('Млрд $')
plt.show()
```

```python
# Столбчатые диаграммы
movies = ['Энни Холл', 'Бен Гур', 'Ксабланка', 'Ганди', 'Вестсайдская история']
num_oscars = [5, 11, 3, 8, 10]

# ширина столбцов по умолчанию 0.8, поэтому добавим 0.1 к левым координатам, 
# чтобы кажды столбец был по центру интервала
xs = [i + 0.1 for i, _ in enumerate(movies)]

# строю столбцы с левыми X-координатами (xs) и высотой (num_oscars)
plt.bar(xs, num_oscars)

plt.ylabel('Количество наград')
plt.title('Мои любимые фильмы')

# добавить метки на оси X с названиями фильмов в центре каждого интервала
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
# линейные графики
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32,16, 8, 4, 2, 1]

# суммарная ошибка
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# метод plt.plot можно вызывать много раз, чтобы показать несколько графиков на одной и той же диаграмме
# зеленая сплошная линия
plt.plot(xs, variance, 'g-', label='дисперсия')
# красная штрихпунктирная
plt.plot( xs, bias_squared, 'r-.', label='смещение^2')
# синяя пунктирная
plt.plot(xs, total_error, 'b:', label='суммарная ошибка')

# если для каждой линии задано название label, то легенда будет показана автоматически
# loc=9 значит наверху посередине
plt.legend(loc=9)
plt.xlabel('Сложность модели')
plt.title('Компромисс между смещение и дисперсией')
plt.show()
```
